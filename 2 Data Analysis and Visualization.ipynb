```python
days=list(range(1,5))
celsius_min=[]
while len(celsius_min)<4:
    degree=float(input("Enter the minimum temperature : "))
    celsius_min.append(degree)
    print(celsius_min)
print('list of minimum temperature : ',celsius_min)
celsius_max=[]
while len(celsius_max)<4:
    degree=float(input('Enter the maximum temperature : '))
    celsius_max.append(degree)
    print(celsius_max)
print('list of maximum temperature : ',celsius_max)
plt.title('Temperature VS Day')
plt.xlabel('Day')
plt.ylabel('Temperature (Degree Celsius)')
#plt.xlim(1,4)   limit of the x axis defined
#plt.ylim(20,30)    limit of the y axis defined
plt.plot(days,celsius_min,
        days,celsius_min,'.g',
         days,celsius_max,
         days,celsius_max,'.r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=Axes3D(fig)
X=np.arange(-4,4,0.25)
Y=np.arange(-4,4,0.25)
X,Y=np.meshgrid(X,Y)
R=np.sqrt(X**2+Y**2)
Z=np.sin(R)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.cm.inferno)
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap=plt.cm.cool)
ax.set_zlim(-2,2)
plt.show
```

```python
sns.lmplot(x='Attack',y='Defense',data=df)
```

```python
sns.lmplot(x='Attack',y='Defense',data=df,
          fit_reg=False,
          hue='Type 1')
```
