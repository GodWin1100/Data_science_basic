```python
menu_data=pd.read_csv('menu.csv')
menu_data.head()
```

```python
melted_df=pd.melt(menu_data,
                 id_vars=['Category','Item','Serving Size','Calories'],
                 var_name='Stat')
melted_df.head()
```

```python
print('Shape of original data',menu_data.shape)
print('Shape of melted data',melted_df.shape)
```

```python
melted_df.Stat.unique()
```

```python
melted_df.Stat.nunique()
```

```python
menu_data.describe()
```

```python
melted_df.describe()
```
